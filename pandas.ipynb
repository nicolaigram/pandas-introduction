## Example

### Loading library and data

```python
import pandas as pd
```

```python
# Creating a DataFrame
# A DataFrame is the main object in the Pandas library

df = pd.read_csv('./ramen-ratings.csv') # Data from https://www.kaggle.com/residentmario/ramen-ratings
df
```

### Exploring data

```python
# List columns
df.columns
```

```python
# See types of data
df.dtypes
```

```python
# Describe data
df.describe()
```

```python
# Show only one columns
df["Stars"]
```

```python
# Show only some columns
df[["Stars", "Style"]]
```

```python
# Filter rows
df.loc[df["Stars"] > 4.5]
```

```python
# Filter rows (multiple conditions)
df.loc[(df["Stars"] > 4.5) & (df["Country"] == "Japan")] # OBS: Remember paranthesis around conditions 
```

```python
# Filter by chicken
df.loc[df["Variety"] == "Chicken"] # Problem: Only has exact matches
```

```python
# Filter by chicken (using Python)
df.loc[df["Variety"].str.lower().str.contains("chicken")] # Solution: Use python to match all cases of chicken
```

```python
# Plot Stars (rating)
df.plot.hist(y="Stars")
```

```python
# Sort by highest rating and show top 5
df.sort_values(by="Stars").tail(5)
```

```python
# Look at single column
df.groupby("Style").size()
```

```python
# Plot single column
df.groupby("Style").size().plot.bar()
```

## Saving data

```python
# Saving a dataframe
df.head(5).to_csv("first5rows.csv")
```

```python
# Saving a plot
df.plot.hist(y="Stars").get_figure().savefig("ratings.png")
```

## Analysing data

Which Style of ramen scores the highest?

```python
# Get the average (mean) of each Style of ramen
style = df.groupby("Style").mean().sort_values(by="Stars")[["Stars"]]                                            
style
```

```python
style.plot.bar()
```

```python
# Expose noise from low amount of ratings
style["Count"] = df.groupby("Style").size()   # Add new column Count
style.sort_values(by="Count")                 # Print sorted by Count
```

```python
# New plot without noise
style.loc[style["Count"] >= 50].plot.bar(y="Stars")
```
